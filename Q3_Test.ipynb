```python
import pandas as pd

coffee_df = pd.read_csv('coffee.csv')

coffee_df
```

```python
import pandas as pd

total_order_df = coffee_df.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

total_order_df = pd.DataFrame(total_order_df)

# Rename columns
total_order_df.columns = ['category', 'product', 'total_order']

total_order_df
```

```python
import matplotlib.pyplot as plt
#import seaborn as sns

# Plot the total orders for each product
plt.figure(figsize=(9, 7))
plt.barh(total_order_df['product'], total_order_df['total_order'])

# Adding labels and title
plt.xlabel('Total Orders')
plt.ylabel('Product')
plt.title('Total Orders by Product Type')

# Show the plot
plt.show()
```
