```python
import pandas as pd

spotify = pd.read_csv("spotify.csv")

spotify
```

```python
spotify['track_album_release_date'] = pd.to_datetime(
    spotify['track_album_release_date'], 
    errors='coerce',           # Coerce invalid dates to NaT
    infer_datetime_format=True
)

# Extract the year from the 'track_album_release_date' column
spotify['track_album_release_year'] = spotify['track_album_release_date'].dt.year

# Filter the DataFrame to include only rows where the release year is 2017
spotify_2017 = spotify[spotify['track_album_release_year'] == 2017]

# Check the resulting DataFrame
print(spotify_2017)
```

```python
def calculate_averages(group):
    avg_loudness = group['loudness'].mean()
    avg_speechiness = group['speechiness'].mean()
    avg_acousticness = group['acousticness'].mean()
    avg_instrumentalness = group['instrumentalness'].mean()
    avg_liveness = group['liveness'].mean()
    avg_valence = group['valence'].mean()
    avg_tempo = group['tempo'].mean()
    avg_duration_minutes = group['duration_ms'].mean() / 60000  # Convert from ms to minutes

    return pd.Series({
        'average_loudness': avg_loudness,
        'average_speechiness': avg_speechiness,
        'average_acousticness': avg_acousticness,
        'average_instrumentalness': avg_instrumentalness,
        'average_liveness': avg_liveness,
        'average_valence': avg_valence,
        'average_tempo': avg_tempo,
        'average_duration(minutes)': avg_duration_minutes
    })

genre_stats = spotify_2017.groupby('playlist_genre').apply(calculate_averages).reset_index()

spotify_df = pd.DataFrame(spotify_2017)
print(spotify_df)

# Save to CSV
genre_stats.to_csv("spotify_genre_averages_2017.csv", index=False)
print("Averages spotify_genre_averages_2017.csv")
```

```python
#Additional

columns_factors = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
spotify_relevant = spotify_2017[columns_factors]

# Calculate correlation matrix
correlation_matrix = spotify_relevant.corr()

print("Correlation Matrix:")
print(correlation_matrix)

#Correlations with 'danceability' and 'energy'
danceability_correlation = correlation_matrix['danceability']
energy_correlation = correlation_matrix['energy']

print("\nCorrelation with Danceability:")
print(danceability_correlation)

print("\nCorrelation with Energy:")
print(energy_correlation)
```
